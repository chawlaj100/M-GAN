## Performing Image Augmentation

```python
import os
import numpy as np
import cv2
import imutils
import matplotlib.pyplot as plt
from PIL import Image
```

```python
path = "CUB_200/images/"
```

```python
def image_augments(p):
    images = []
    
    img = cv2.imread(p)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    images.append(img_rgb)
    
    img_h = cv2.flip(img_rgb, 1)
    images.append(img_h)
    
    a1 = np.random.randint(0, 30)
    img_r = imutils.rotate(img_rgb, a1)
    images.append(img_r)
    
    a2 = np.random.randint(-30, -1)
    img_ro = imutils.rotate(img_rgb, a2)
    images.append(img_ro)
    
    
    plt.figure(figsize=(15, 10))
    for index, image in enumerate(images):
        plt.subplot(1, 4, index + 1)
        plt.imshow(image)
        plt.axis("off")
    plt.show()
```

```python
temp = path + "American Crow/American_Crow_0001_25053.jpg"
```

```python
image_augments(temp)
```

```python
folders = os.listdir(path)
```

```python
folder = folders[0:69]
folder
```

```python
i = 0
for f in folder:
    images = os.listdir(path+f)
    for im in images:
        temp = path+f+"/"+im
        img = cv2.imread(temp)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        t1 = path+f+"/h_"+im
        img_h = cv2.flip(img_rgb, 1)
        pil_h = Image.fromarray(img_h)
        pil_h.save(t1)
        
        t2 = path+f+"/r_"+im
        a1 = np.random.randint(0, 30)
        img_r = imutils.rotate(img_rgb, a1)
        pil_r = Image.fromarray(img_r)
        pil_r.save(t2)
        
        t3 = path+f+"/ro_"+im
        a2 = np.random.randint(-30, -1)
        img_ro = imutils.rotate(img_rgb, a2)
        pil_ro = Image.fromarray(img_ro)
        pil_ro.save(t3)
    i += 1
    if i % 10 == 0:
        print("Done: ",i)
```
